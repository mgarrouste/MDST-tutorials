<h2> Cleanup (optional)</h2>

_Use this space to rename columns, deal with missing data, etc._

```python
df.isna().sum()
```

```python
df.count()
```

```python
df.rename({
    'GRADES_PK_G':'enroll_kinder',
    'GRADES_KG_G':'enroll_kinder',
    'GRADES_4_G':'enroll_4',
    'GRADES_8_G':'enroll_8',
    'GRADES_12_G':'enroll_12',
    'GRADES_1_8_G':'enroll_primary',
    'GRADES_9_12_G':'enroll_hs',
    'GRADES_ALL_G':'enroll_all',
    'ENROLL':'ENROLL_ALL_EST'
    },
    axis=1,inplace=True)
def standardize_names(df):
    df.rename(mapper = str.lower, axis = 1, inplace=True)
    return df
df = standardize_names(df)
df.head()
```

```python
#Dropping the rows without data about the math tests scores in 8th grade
df.dropna(subset = ['avg_math_8_score'], inplace = True)
df.count()
```

```python
#Looking at the estimated and recorded enrollment to see if one can be used to fill the other
((df.enroll_all_est - df.enroll_all)/df.enroll_all).describe()
```

```python
#The standard deviation is not so high so we can fill a couple of missing estimates, 
#the rest is taken care of with the mean
df.enroll_all_est = df.enroll_all_est.fillna(df.enroll_all)
df.enroll_all_est = df.enroll_all_est.fillna(df.enroll_all_est.mean())
df.count()
```

```python
#Missing values in the expenditure: fill the 'other expenditure', which has more missing values, 
#with the total minus all the others
df.other_expenditure = df.other_expenditure.fillna(df.total_expenditure - df.instruction_expenditure 
                                                   - df.support_services_expenditure 
                                                   - df.capital_outlay_expenditure)
#Rest of missing values with mean
df.other_expenditure = df.other_expenditure.fillna(df.other_expenditure.mean())
df.instruction_expenditure = df.instruction_expenditure.fillna(df.instruction_expenditure.mean())
df.support_services_expenditure = df.support_services_expenditure.fillna(df.support_services_expenditure.mean())
df.capital_outlay_expenditure = df.capital_outlay_expenditure.fillna(df.capital_outlay_expenditure.mean())
#Total expenditure missing values as the sum of the others so that it still has the same meaning
df.total_expenditure = df.total_expenditure.fillna(df.instruction_expenditure +
                                                  df.support_services_expenditure+
                                                  df.capital_outlay_expenditure+
                                                  df.other_expenditure)
df.count()
```

```python
#Missing values in the revenue data: fill the state, federal and local revenues missing values 
#with the mean of the other values and sum them up for the total revenus missing values
df.federal_revenue = df.federal_revenue.fillna(df.federal_revenue.mean())
df.state_revenue = df.state_revenue.fillna(df.state_revenue.mean())
df.local_revenue = df.local_revenue.fillna(df.local_revenue.mean())
df.total_revenue = df.total_revenue.fillna(df.local_revenue +
                                          df.state_revenue +
                                          df.federal_revenue)
df.count()
```

<h2> Feature Selection </h2>

_Use this space to modify or create features_

```python
df['total_revenue_per_student'] = df['total_revenue']/df['enroll_all_est']
df['local_revenue_per_student'] = df['local_revenue']/df['enroll_all_est']
df['total_expenditure_per_student'] = df['total_expenditure']/df['enroll_all_est']
df['instruction_expenditure_per_student'] = df['instruction_expenditure']/df['enroll_all_est']
```

Final feature list: total_revenue_per_student,  local_revenue_per_student, total_expenditure_per_student, instruction_expenditure_per_student,  avg_math_8_score, year, state

Feature selection justification: I want to see the influence of the revenue and expenditure on the average math score in 8th grade.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['local_revenue_per_student','total_revenue_per_student',
        'year']]
y = df.loc[X.index]['avg_math_8_score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.2, random_state=0)
```

<h2> Prediction </h2>

Chosen ML task: Regression

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
print(model.intercept_)
print(model.coef_)
```

```python
y_pred = model.predict(X_test)
model.score(X_test, y_test)
```

Not very good at predicting anything...

```python
# FOR REGRESSION ONLY: (pick a single column to visualize results)

# Results from this graph _should not_ be used as a part of your results -- it is just here to help with intuition. 
# Instead, look at the error values and individual intercepts.


col_name = 'total_revenue_per_student'
col_index = X_train.columns.get_loc(col_name)

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
plt.scatter(X_test[col_name], model.predict(X_test), color = "blue")

new_x = np.linspace(X_train[col_name].min(),X_train[col_name].max(),200)
intercept = model.predict([X_train.sort_values(col_name).iloc[0]]) - X_train[col_name].min()*model.coef_[col_index]
plt.plot(new_x, intercept+new_x*model.coef_[col_index])

plt.legend(['controlled model','true training','predicted training','predicted testing'])
plt.xlabel(col_name)
plt.ylabel('Instruction expenditure per student')
```

<h2> Classification </h2>

```python
#Best students in math
y = df.loc[X.index]['avg_math_8_score']>285
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=0)
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
# FOR CLASSIFICATION ONLY:
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(model, X_test, y_test,
                         cmap=plt.cm.Blues)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score
accuracy_score(y_test, model.predict(X_test))
```

Better!

```python
precision_score(y_test, model.predict(X_test))
```

```python
recall_score(y_test, model.predict(X_test))
```
